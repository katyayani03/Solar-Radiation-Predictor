```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
diffuse_path = "diffuse.csv"
global_path = "global.csv"
df = pd.read_csv(diffuse_path)
df.name = "DIFFUSE"
df2 = pd.read_csv(global_path)
df2.name = "GLOBAL"
```

```python
def eda(dataframe):
  dataframe.info()
  print("\n")
  unique_counts = dataframe.groupby(list(dataframe.INDEX)).size().reset_index(name='Count')
  print("Count of data entry for each station \n", unique_counts)
  return dataframe.head()
```

```python
eda(df)
```

```python
eda(df2)
```

```python
stations = {42339: "Jodhpur",
            42647: "Ahmedabad",
            42667: "Bhopal",
            42701: "Ranchi",
            42809: "Kolkata",
            43003 : "Mumbai",
            43185: "Machlipatnam",
            43284: "Mangalore"}

def insert_station_name(dataframe):
  dataframe['STATION'] = dataframe['INDEX'].apply(lambda x: stations.get(x))
  st = dataframe.pop('STATION')
  dataframe.insert(1, 'STATION', st)
  return dataframe.head()
```

```python
insert_station_name(df)
```

```python
insert_station_name(df2)
```

```python
def format_date(dataframe):
  # df["DATE"] = "df["DT"].astype(str) + "-" + df["MN"].astype(str) + "-" + df["YEAR"].astype(str)
  dataframe["DATE"] = '15' + "-" + dataframe["MN"].astype(str) + "-" + dataframe["YEAR"].astype(str)
  date = dataframe.pop('DATE')
  dataframe.insert(1, 'DATE', date)
  dataframe['DATE'] = pd.to_datetime(dataframe['DATE'], format='%d-%m-%Y', errors='coerce')
  dataframe.drop(columns=["YEAR", "MN", "INDEX"], inplace=True)
  print(dataframe.info())
  return dataframe.head()
```

```python
format_date(df)
```

```python
format_date(df2)
```

```python
def melt_dataframe(dataframe):
  dataframe = dataframe.melt(id_vars=['DATE', 'STATION'],
             value_vars=[f"{hour}HR"  for hour in range(5, 21)],
             var_name='Hour',
             value_name = (f'{dataframe.name} MJ/m^2'),
             )
  dataframe.info()
  return dataframe
def format_dataframe(dataframe):
  dataframe['Hour'] = dataframe['Hour'].str.replace('HR', '').astype(int)
  dataframe['DATETIME'] = dataframe['DATE'] + pd.to_timedelta(dataframe['Hour'], unit='h')
  dataframe.drop(['DATE', 'Hour'], axis=1, inplace=True)
  datetime = dataframe.pop('DATETIME')
  dataframe.insert(1, 'DATETIME', datetime)

  # dataframe = dataframe.groupby('STATION').apply(lambda x: x.sort_values('DATETIME')).reset_index(drop=True)

  dataframe.info()
  return dataframe.head()
```

```python
df = melt_dataframe(df)
```

```python
df.head()
```

```python
format_dataframe(df)
df = df.groupby('STATION').apply(lambda x: x.sort_values('DATETIME')).reset_index(drop=True)
```

```python
df2 = melt_dataframe(df2)
```

```python
df2.head()
```

```python
format_dataframe(df2)
df2 = df2.groupby('STATION').apply(lambda x: x.sort_values('DATETIME')).reset_index(drop=True)
```
